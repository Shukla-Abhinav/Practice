# Two Pointer Approach 
## Check if a string is palindrome or not using 2 pointer approach

```python
def isPlaindrome(s):
    i = 0    # from starting
    j = len(s) - 1    #from end
    
    while i < j:
        if s[i] != s[j]:
            return False
        i += 1
        j -= 1
        
    return True
```

```python
# Calling the function for multiple testcases
t = int(input())
    for i in range(t):
        s= input()
        print(isPlaindrome(s))
```

# Find Elements from a sorted array equal to a target value

```python
def findTarget(a,target):
    i = 0
    j = len(a)-1
    while i < j:
        if a[i] + a[j] == target:
            print(i,'+',j,'=',target)
            print('Index :',i,j)
            return True
        elif a[i] + a[j] < target:
            i += 1
        else:
            j -= 1
    return False
```

```python
a = [4,5,3,6,7,2]
target = 9
findTarget(a,target)
```

# Array pair sum

```python
def findTargetPair(a,k):
    st = set()
    output = []
    for i in range(len(a)):
        complement = k - a[i]
        if complement in st:
            output.append([complement,a[i]])
        else:
            st.add(a[i])
    return output
```

```python
a = [4,5,6,9,8,7,5]
k = 15
findTargetPair(a,k)
```
